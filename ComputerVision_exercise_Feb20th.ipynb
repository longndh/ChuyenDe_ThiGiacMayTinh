```python
#Nhận diện khuôn mặt và lưu từng khuôn mặt ra một file mới

import numpy as np
import cv2 as cv

face_cascade = cv.CascadeClassifier('haarcascade_frontalface_default.xml')

img = cv.imread('beatles.jpg')
gray = cv.cvtColor(img, cv.COLOR_RGB2GRAY)

faces = face_cascade.detectMultiScale(gray, 1.2, 5)

i = 0
for (x,y,w,h) in faces:
    crop_img = img[y:y+h, x:x+w]
    i = i + 1
    file_output = 'beatles_face' + str(i) + '.jpg'
    cv.imwrite(file_output, crop_img)
```

```python
#Hiển thị ảnh trên jupyter notebook (thực hiện sau đoạn code trên)
from matplotlib import pyplot as plt

im_cvt = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.rcParams["figure.figsize"] = (6, 6)
plt.imshow(im_cvt)
plt.show()

for j in range (1, i+1):
    file_name = 'beatles_face' + str(j) + '.jpg'
    im = cv.imread(file_name)
    im_cvt = cv.cvtColor(im, cv.COLOR_BGR2RGB)
    plt.rcParams["figure.figsize"] = (2, 2)
    plt.imshow(im_cvt)
    plt.show()
```
